```python
import pandas
import pandas as pd
import sklearn.model_selection
from sklearn.preprocessing import OneHotEncoder
import matplotlib.pyplot as plt

data = pandas.read_csv("StudentPerformanceFactorsPuhas.csv", sep=";")

data
```

```python
data = pandas.get_dummies(data, columns=data.columns[(data.dtypes == 'object') | (data.dtypes == 'category')])
data
```

```python
x = data.drop("Exam_Score", axis=1)
y = data["Exam_Score"]
print(x.shape,
y.shape)
X_train_val, X_test, y_train_val, y_test = sklearn.model_selection.train_test_split(x, y, train_size=0.8)
X_train, X_val, y_train, y_val = sklearn.model_selection.train_test_split(X_train_val, y_train_val, train_size=0.8)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

model = KNeighborsRegressor(n_neighbors=10, metric="euclidean")
model.fit(X_train, y_train)
accuracy = mean_squared_error(y_test, model.predict(X_test))
accuracy
```

```python
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor(criterion="squared_error")
model.fit(X_train, y_train)
accuracy = mean_squared_error(y_test, model.predict(X_test))
accuracy
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(X_train, y_train)
accuracy = mean_squared_error(y_val, model.predict(X_val))
accuracy
```

```python
from sklearn.svm import SVR

model = SVR(kernel="linear", degree=2, gamma="auto")
model.fit(X_train[["Attendance", "Hours_Studied", "Previous_Scores", "Tutoring_Sessions", "Sleep_Hours", "Physical_Activity"]], y_train)

accuracy = mean_squared_error(y_val, model.predict(X_val[["Attendance", "Hours_Studied", "Previous_Scores", "Tutoring_Sessions", "Sleep_Hours", "Physical_Activity"]]))
display(accuracy)

import pickle

with open("model.pkl", "wb") as file:
    pickle.dump(model, file)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import pandas as pd

model = RandomForestRegressor()
model.fit(X_train, y_train)
pred = model.predict(X_val)
mse = mean_squared_error(y_val, pred)
print(f'Mean Squared Error: {mse}')

importances = model.feature_importances_
feature_importance_df = pd.DataFrame({"Feature": X_train.columns, "Importance": importances})
print(feature_importance_df.sort_values(by="Importance"))
```
